```python
#matches against different oppostions plot
fig = plt.figure(figsize=[10,7])
plt.pie(opp_freq, labels = opp_freq.index)
```

```python
#matches played at different position
fig = plt.figure(figsize=[10,7])
plt.pie(df["Ground"].value_counts(), labels = df["Ground"].value_counts().index)
```

```python
#all the matches where virat scored a century
century = df[df["Runs"]>=100]
```

```python
century
```

```python
#the matches where the strike rate was greater than 100
strike_rate = df[df["SR"]>=100]
```

```python
strike_rate
```

```python
#all the matches where virat played with srilanka and scored a century
cent = df[df["Opposition"]=="v Sri Lanka"]
c = cent[cent["Runs"]>=100]
```

```python
c
```

```python
#grading the runs, if scored century and more then OG else noob

def find_centuries(x):
    if x>=100:
        return "OG"
    else:
        return "noob"
df["Centuries"] = df["Runs"].apply(find_centuries)
```

```python
df
```

```python
#6's against each oppositions
six = df[df["6s"]>=1]
```

```python
six
```

```python
fig = plt.figure(figsize=[12,7])
plt.bar(six["Opposition"], six["6s"])
```

```python
#4's against each oppositions
four = df[df["4s"]>=1]
```

```python
four
```

```python
fig = plt.figure(figsize=[15,10])
plt.bar(four["Opposition"], four["6s"])
```

```python
#runs scored against different country
runs = df.groupby("Opposition")["Runs"].sum()

fig = plt.figure(figsize=[15,10])
plt.bar(runs.index, runs.values, width = .2)
```

```python
#maximum runs against each opposition
mx = df.groupby("Opposition")["Runs"].max()
```
